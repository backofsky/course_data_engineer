## Группировка вакансий по направлениями по направлениями (DS, DE, Software Engenering, etc.) 

```python
# Сгрупируем по столбцу 'job_title' и выведем первые 5 значений
df.groupby('job_title').count().head()
```

## Средняя и медианная зарплата по группам вакансий

```python
# Выведем среднюю и медианное значения для 'salary_in_usd'
df.groupby('job_title')['salary_in_usd'].agg(['mean', 'median'])
```

## Средняя и медианная зарплата по каждому региону

```python
# Выведем среднюю и медианное значения 'salary_in_usd' по каждому региону
df.groupby('company_location')['salary_in_usd'].agg(['mean', 'median'])
```

```python
# Выведем среднюю и медианное значения 'salary_in_usd' по каждому региону и группам вакансий
df.groupby(['job_title', 'company_location'])['salary_in_usd'].agg(['mean', 'median'])
```

## Самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат

```python
# Выведем самую высокооплачиваемую вакансию
df[df['salary_in_usd'] == df.groupby('job_title')['salary_in_usd'].mean().max()]
```

Таким образом, *Data Analytics Lead - самая высокооплачиваемая вакансия*.

## Процентное соотношение каждого региона по вакансиям от всех вакансий

```python
# Всего вакансий
sum_vac = df.groupby('company_location')['salary_in_usd'].count().sum()

# Количество вакансий в каждом регионе
count_vac = df.groupby('company_location')['salary_in_usd'].count()

# Процентное соотношение
count_vac * 100 / sum_vac
```

## Корреляция уровня опыта от зарплаты

**Уровень опыта:**

EN => Entry-level or Junior 

MI => Mid-level or Intermediate 

SE => Senior-level or Expert 

EX => Executive-level or Director

```python
# Размер графика
plt.figure(figsize=(12, 6))

# График
plt.bar(df['experience_level'], df['salary_in_usd'], color ='maroon', width=0.8)

# Название графика
plt.title("Salaries on the basis of Experience level")

# Подпись оси Х
plt.xlabel("Experience Level")

# Подпись оси Y
plt.ylabel("Salary (USD)")

# Отображение
plt.show()
```

*Наблюдаем взаимосвязь между 'experience_level' и 'salary_in_usd'. Видим корреляцию между параметрами.*
