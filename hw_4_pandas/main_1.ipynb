```python
# Импортируем нужные библиотеки

# Pandas
import pandas as pd

# Для работы с JSON
import json
```

```python
# Чтение данных
with open('./data_01102022.json', 'r', encoding='utf-8') as file:
    data = json.loads(file.read())
```

```python
# Упорядоченный список значений
for elem in data['data']:
    print(elem)
```

Видим, что в столбцах 'work experience' и 'salary' есть нечисловые значения. Очистим значения 'work experience' и 'salary', а также 'region'.

Усредним значения 'work experience' и 'salary'.

```python
# Предварительно очистим список, который мы можем обработать в Pandas
data_list = []

for elem in data['data']:
    work_list = []
    salary_list = []
    region_list = []
    # Опыт работы      
    work_exp = elem['work experience'].replace('не требуется', '0').replace('более 6', '7').replace('–', ' ').replace('года', '').replace('лет', '').split(' ')
    try:
        work_list.append(int(work_exp[0]) + int(work_exp[1]) / len(work_exp))
    except:
        work_list.append(int(work_exp[0]))
    # Зараплата
    try:
        salary = elem['salary'].replace(u'\xa0', '').split('-')
        salary_list.append(int(salary[0]) + int(salary[1]) / len(salary))
    except:              
        salary_list.append(None)     
    # Регион  
    region_list = elem['region'].split(',')[0]
    
    data_list.append([elem['title'], *work_list, *salary_list, region_list])
```

```python
# Подгрузка списка в таблицу pandas
df = pd.DataFrame(data_list, columns=data['data'][0].keys())
```

## Группировка вакансий по направлениями по направлениями (DS, DE, Software Engenering, etc.) 

```python
# Сгрупируем по столбцу 'title' и выведем первые 5 значений
df.groupby('title').count().head()
```

## Средняя и медианная зарплата по группам вакансий

```python
# Выведем среднюю и медианное значения для 'salary'
df.groupby('title')['salary'].agg(['mean', 'median'])
```

## Средняя и медианная зарплата по каждому региону

```python
# Выведем среднюю и медианное значение 'salary' по каждому региону
df.groupby('region')['salary'].agg(['mean', 'median'])
```

```python
# Выведем среднюю и медианное значение 'salary' по каждому региону и группам вакансий
df.groupby(['title', 'region'])['salary'].agg(['mean', 'median'])
```

## Самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат

```python
# Выведем самую высокооплачиваемую вакансию
df[df['salary'] == df.groupby('title')['salary'].mean().max()]
```

Таким образом, *Программист С++ Senior и Senior Software Engineer - самые высокооплачиваемые вакансии*.

## Процентное соотношение каждого региона по вакансиям от всех вакансий

```python
# Всего вакансий
sum_vac = df.groupby('region')['salary'].count().sum()

# Количество вакансий в каждом регионе
count_vac = df.groupby('region')['salary'].count()

# Процентное соотношение
count_vac * 100 / sum_vac
```
